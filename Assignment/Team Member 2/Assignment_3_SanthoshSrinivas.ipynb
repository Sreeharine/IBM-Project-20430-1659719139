```python
#Label Encoding for categorical Values
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['Sex']= label_encoder.fit_transform(df['Sex'])
df.head()
```

```python
q1=df["Length"].quantile(0.10)
q2=df["Length"].quantile(0.90)
df["Length"] = np.where(df["Length"] <q1, q1,df["Length"])
df["Length"] = np.where(df["Length"] >q2, q2,df["Length"])
print(df["Length"].skew())

q1=df['Diameter'].quantile(0.10)
q2=df['Diameter'].quantile(0.90)
df['Diameter'] = np.where(df['Diameter'] <q1, q1,df['Diameter'])
df['Diameter'] = np.where(df['Diameter'] >q2, q2,df['Diameter'])
print(df['Diameter'].skew())

q1=df['Height'].quantile(0.10)
q2=df['Height'].quantile(0.90)
df['Height'] = np.where(df['Height'] <q1, q1,df['Height'])
df['Height'] = np.where(df['Height'] >q2, q2,df['Height'])
print(df['Height'].skew())

q1=df['Whole weight'].quantile(0.10)
q2=df['Whole weight'].quantile(0.90)
df['Whole weight'] = np.where(df['Whole weight'] <q1, q1,df['Whole weight'])
df['Whole weight'] = np.where(df['Whole weight'] >q2, q2,df['Whole weight'])
print(df['Whole weight'].skew())

q1=df['Shucked weight'].quantile(0.10)
q2=df['Shucked weight'].quantile(0.90)
df['Shucked weight'] = np.where(df['Shucked weight'] <q1, q1,df['Shucked weight'])
df['Shucked weight'] = np.where(df['Shucked weight'] >q2, q2,df['Shucked weight'])
print(df['Shucked weight'].skew())

q1=df['Viscera weight'].quantile(0.10)
q2=df['Viscera weight'].quantile(0.90)
df['Viscera weight'] = np.where(df['Viscera weight'] <q1, q1,df['Viscera weight'])
df['Viscera weight'] = np.where(df['Viscera weight'] >q2, q2,df['Viscera weight'])
print(df['Viscera weight'].skew())


q1=df['Shell weight'].quantile(0.10)
q2=df['Shell weight'].quantile(0.90)
df['Shell weight'] = np.where(df['Shell weight'] <q1, q1,df['Shell weight'])
df['Shell weight'] = np.where(df['Shell weight'] >q2, q2,df['Shell weight'])
print(df['Shell weight'].skew())

q1=df['Rings'].quantile(0.10)
q2=df['Rings'].quantile(0.90)
df['Rings'] = np.where(df['Rings'] <q1, q1,df['Rings'])
df['Rings'] = np.where(df['Rings'] >q2, q2,df['Rings'])
print(df['Rings'].skew())
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.svm import SVR
regrassor = SVR(kernel = 'rbf')
regrassor.fit(X_train, y_train)
```

```python
y_pred = regrassor.predict((X_test))
df3 = pd.DataFrame({'Predicted value': y_pred, 'Real Value': y_test})
df3
```

```python
from sklearn.metrics import mean_squared_error
import math
print(mean_squared_error(y_test, y_pred))
print(math.sqrt(mean_squared_error(y_test, y_pred)))
```
