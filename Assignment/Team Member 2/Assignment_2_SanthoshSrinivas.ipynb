#ENCODING

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
dataframe['Gender']= label_encoder.fit_transform(dataframe['Gender'])
dataframe['Geography']= label_encoder.fit_transform(dataframe['Geography'])
dataframe
```

```python
#Outiers Identification And Replacing using Quantile-based Flooring and Capping for continuous variables
df2=dataframe[['CustomerId','CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary']]

q1=dataframe["CustomerId"].quantile(0.10)
q2=dataframe["CustomerId"].quantile(0.90)
dataframe["CustomerId"] = np.where(dataframe["CustomerId"] <q1, q1,dataframe["CustomerId"])
dataframe["CUstomerId"] = np.where(dataframe["CustomerId"] >q2, q2,dataframe["CustomerId"])
print(dataframe["CustomerId"].skew())

q1=dataframe['CreditScore'].quantile(0.10)
q2=dataframe['CreditScore'].quantile(0.90)
dataframe['CreditScore'] = np.where(dataframe['CreditScore'] <q1, q1,dataframe['CreditScore'])
dataframe['CreditScore'] = np.where(dataframe['CreditScore'] >q2, q2,dataframe['CreditScore'])
print(dataframe['CreditScore'].skew())

q1=dataframe['Age'].quantile(0.10)
q2=dataframe['Age'].quantile(0.90)
dataframe['Age'] = np.where(dataframe['Age'] <q1, q1,dataframe['Age'])
dataframe['Age'] = np.where(dataframe['Age'] >q2, q2,dataframe['Age'])
print(dataframe['Age'].skew())

q1=dataframe['Tenure'].quantile(0.10)
q2=dataframe['Tenure'].quantile(0.90)
dataframe['Tenure'] = np.where(dataframe['Tenure'] <q1, q1,dataframe['Tenure'])
dataframe['Tenure'] = np.where(dataframe['Tenure'] >q2, q2,dataframe['Tenure'])
print(dataframe['Tenure'].skew())

q1=dataframe['Balance'].quantile(0.10)
q2=dataframe['Balance'].quantile(0.90)
dataframe['Balance'] = np.where(dataframe['Balance'] <q1, q1,dataframe['Balance'])
dataframe['Balance'] = np.where(dataframe['Balance'] >q2, q2,dataframe['Balance'])
print(dataframe['Balance'].skew())

q1=dataframe['NumOfProducts'].quantile(0.10)
q2=dataframe['NumOfProducts'].quantile(0.90)
dataframe['NumOfProducts'] = np.where(dataframe['NumOfProducts'] <q1, q1,dataframe['NumOfProducts'])
dataframe['NumOfProducts'] = np.where(dataframe['NumOfProducts'] >q2, q2,dataframe['NumOfProducts'])
print(dataframe['NumOfProducts'].skew())


q1=dataframe['EstimatedSalary'].quantile(0.10)
q2=dataframe['EstimatedSalary'].quantile(0.90)
dataframe['EstimatedSalary'] = np.where(dataframe['EstimatedSalary'] <q1, q1,dataframe['EstimatedSalary'])
dataframe['EstimatedSalary'] = np.where(dataframe['EstimatedSalary'] >q2, q2,dataframe['EstimatedSalary'])
print(dataframe['EstimatedSalary'].skew())
```
